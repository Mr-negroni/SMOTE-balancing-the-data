```python
from sklearn.datasets import make_classification
X,y = make_classification(n_classes=2, class_sep=0.5, weights=[0.05, 0.95], n_informative=2,
                          n_redundant=0, flip_y=0, n_samples=1000, random_state=10,
                          n_clusters_per_class=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(X[:,0],X[:,1],c=y)
plt.show()
```

```python
import pandas as pd
data = pd.DataFrame(X)
data['y'] = y
```

```python
data['y'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res,y_res = sm.fit_resample(X,y)
```

```python
y_res.shape
```

```python
X_res.shape
```

```python
smote_df = pd.DataFrame(X_res)
smote_df['y'] = y_res
```
